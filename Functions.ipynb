#### **4. What is the purpose of the `return` statement in a Python function?**

- The `return` satament in a python function serves to send a value back to the caller of the function.
- When a `return` statment is encountered, the function immediately stops executing, and the value specified after the `return` is returned to the point where the function was called.
- If function doesn't have a return statement, it implicitly returns `None` by default. Given below example proves this statment

#### **5. What are iterators in Python and how do they differ from iterables?**

1. Iterable
  - It is an object capable of returning its members one by one. Examples include lists, tuples, strings, and dictionaries. We can loop over iterables.
  - An object is iterable if it has an `__iter__` method and `__next__` in `dir()` of the object.

2. Iterator
  - It is an object that represents a stream of data. It's what actually does the iterating.
  - It also keeps track of the state of the iteration and knows how to get the next element.
  - An iterator object must implement two methods:
      - `iter()`: Returns the iterator object itself.
      - `next()`: Returns the next item from the stream. If there are no more items, it raises a `StopIteration` exception.

#### **6. Explain the concept of generators in Python and how they are defined.**

- Generators are a special type of function in Python that return an iterator. They are defined like normal functions but use the `yield` keyword instead of `return`.
- When a generator function is called, it doesn't execute the code immediately. Instead, it returns a generator object. The code within the generator function is executed only when the `next()` method is called on the generator object.
- The main advantage of generators is that they generate values on the fly, one at a time, as they are requested. This makes them very memory-efficient, especially when dealing with large datasets, as they don't need to store all the values in memory at once.
- Example :

````python
def my_generator(a):
  """A simple generator function"""
  for i in a:
    yield i**2

# Create a generator object
gen = my_generator([3,2,5,3,2])

print(gen)
print(next(gen))
print(next(gen))
print(next(gen))
````

#### **7. What are the advantages of using generators over regular functions**

As i explained earlier, It gives the advantages like,

1. Memory Efficiency :
  - Generators produce items one at a time and on the fly. They don't store all the items in memory at once, making them ideal for working with large datasets or infinite sequences where creating a full list would consume excessive memory.

2. Performance :
  - For large datasets, generating values on demand can be faster than creating and returning an entire list, especially if you only need to process a few items from the sequence.

3. Simplicity and Readability :
  - Generator functions are often more concise and easier to read than writing custom iterator classes with `__iter__` and `__next__` methods. The `yield` keyword makes the code look like a regular function while providing the benefits of iteration.

4. Handling Infinite Sequences :
  - Generators can represent infinite sequences of data because they don't need to generate all values upfront. This is impossible with regular functions that return a finite collection.


#### **8. What is a lambda function in Python and when is it typically used ?**

- A lambda function in Python is a small, anonymous function. It's defined using the `lambda` keyword and can take any number of arguments, but can only have one expression. The expression is evaluated and returned.

- Lambda functions are typically used for short, simple operations where a full function definition using `def` would be lengthy for small functions. They are often used in conjunction with functions that take other functions as arguments, such as `map()`, `filter()`, and `sorted()`.

````python
# Example of a lambda function
add = lambda x, y: x + y
print(add(5, 3))

# Using lambda with map()
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)

# Using lambda with filter()
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)

# Using lambda with sorted()
pairs = [('one'), ('two'), ('three')]
sorted_pair = sorted(pairs, key=lambda pair: pair)
print(sorted_pair)
````

#### **9. Explain the purpose and usage of the `map()` function in Python.**

- The `map()` function in Python is a built-in function that applies a given function to each item of an iterable (like a list or tuple) and returns an iterator that yields the results.
- It's a concise way to perform the same operation on all elements of a sequence without using an explicit loop.

- The syntax of `map()` is:

````python
# Define a function to square a number
def square(x):
  return x * x

# Create a list of numbers
numbers = [1, 2, 3, 4, 5]

# Use map() to apply the square function to each number
squared_numbers_iterator = map(square, numbers)

# Convert the iterator to a list to see the results
squared_numbers = list(squared_numbers_iterator)
print(squared_numbers)

# Using a lambda function with map()
squared_numbers_lambda = list(map(lambda x: x**2, numbers))
print(squared_numbers_lambda)
````

#### **10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

- `map()`: Applies a given function to each item of an iterable and returns an iterator that yields the results. It's used for transforming elements.

- `filter()`: Filters elements from an iterable based on a function that returns `True` or `False`. It returns an iterator that yields only the elements for which the function returned `True`. It's used for selecting elements.

- `reduce()`: Applies a function of two arguments cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value. It's part of the `functools` module. It's used for aggregating elements.

````python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Using map() to square each number
squared_numbers = list(map(lambda x: x**2, numbers))
print(f"map() result    : {squared_numbers}")

# Using filter() to get even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"filter() result : {even_numbers}")

# Using reduce() to find the sum of numbers
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(f"reduce() result : {sum_of_numbers}")
````

````python
# Method 1
def even_sum(x):
  even = []
  for i in x:
    if i % 2 == 0:
      even.append(i)

  return f"Sum of even numbers in the list is : {sum(even)}"
````

#### **4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

````python
def is_prime(num):
  """
  Checks if a number is prime within the range of 1 to 200.
  """
  if num < 2 or num > 200:
    return "Number must be between 1 and 200."
  if num <= 1:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

# Example usage:
print(is_prime(7))
print(is_prime(10))
print(is_prime(199))
print(is_prime(0))
print(is_prime(250))
````

#### **7. Implement a generator function that reads a file line by line and yields each line as a string.**

````python
import os

# Create a dummy file for demonstration if it doesn't exist
file_content = """This is the first line.
This is the second line.
This is the third line.
And the fourth one."""

file_name = "my_example_file.txt"

# Check if the file already exists before creating it
if not os.path.exists(file_name):
    with open(file_name, "w") as f:
        f.write(file_content)
    print(f"Created dummy file: {file_name}")
else:
    print(f"Dummy file already exists: {file_name}")


def read_file_line_by_line(filename):
  """
  A generator function that reads a file line by line and yields each line.
  """
  try:
    with open(filename, 'r') as f:
      for line in f:
        yield line.strip() # Use strip() to remove leading/trailing whitespace, including newline characters
  except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
  except Exception as e:
    print(f"An error occurred: {e}")

# Example usage:
# Iterate through the lines using the generator
print("\nReading file using the generator:")
for line in read_file_line_by_line(file_name):
  print(line)

# Optional: Clean up the dummy file
# os.remove(file_name)
# print(f"\nRemoved dummy file: {file_name}")
````

#### **10. Create a Python program that uses `filter()` to remove all the vowels from a given string**

````python
def Remove_Vowel(string):
  vowels = "AEIOUaeiou"
  # Use a lambda function with filter to keep characters that are not vowels
  result = filter(lambda char: char not in vowels, string)
  # Join the filtered characters back into a string
  return "".join(result)
````

````python
print(Remove_Vowel("Hello World"))
print(Remove_Vowel("Python Programming"))
````
